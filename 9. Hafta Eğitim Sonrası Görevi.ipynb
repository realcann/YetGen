# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time
class Kumanda():
    def __init__(self,tv_durum = "kapalı", tv_ses = 0, kanal_sayısı = 0, kanal_list = ["trt","fox tv","show tv"], kanal = "trt","fox tv","show tv","star tv","kanal D" ):
        self.tv_durum = tv_durum
        self.tv_ses = tv_ses
        self.kanal_sayısı = kanal_sayısı
        self.kanal_list = kanal_list
        self.kanal = kanal

    def tv_ac(self):
        if self.tv_durum == "kapalı":
            print("Televizyon açılıyor...")
            self.tv_durum = "açık"
        else:
            print("Televizyon zaten açık efendim")
    def tv_kapat(self):
        if self.tv_durum == "açık":
            print("Televizyon kapanıyor...")
            self.tv_durum = "kapalı"
        else:
            print("Televizyon zaten kapalı efendim")
    def ses_ayarları(self):
        while True:
            islem = input("ses arttır = >, ses azalt = <, çıkıs = Ç")
            if islem == "<":
                self.tv_ses -=1
                print(f"yeni ses düzeyi {self.tv_ses}")
            elif islem == ">":
                self.tv_ses +=1
                print(f"yeni ses düzeyi {self.tv_ses}")
            else:
                print(f"ses güncellendi {self.tv_ses}")
                break
    def kanal_ekle(self,yeni_kanal):
        print("Kanal ekleniyor...")
        time.sleep(1)
        self.kanal_list.append(yeni_kanal)
        print("Kanal eklendi")
    def rastgele_kanal(self):
        rastgele = random.randint(0,len(self.kanal_list)-1)
        self.kanal = self.kanal_list[rastgele]
        print(f"mevcut kanal = {self.kanal}")
    def __len__(self):
        return len(self.kanal_list)
    def __str__(self):
        return "TV durum = {}\nses durum = {}\nkanal listesi = {}\nşu an ki kanal = {}".format(self.tv_durum, self.tv_ses,self.kanal_list,self.kanal)

kumanda = Kumanda()

print("""Televizyon uygulaması:
1. Televizyonu Aç
2. Televizyonu Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri
""")

while True:
    islem = input("İşlem seçiniz efendim")
    if islem == "q":
        print("İşlem sonlandırılıyor...")
        break
    elif islem == "1":
        kumanda.tv_ac()
    elif islem == "2":
        kumanda.tv_kapat()
    elif islem == "3":
        kumanda.ses_ayarları()
    elif islem == "4":
        kanal_ismi = input("Kanal ismi giriniz efendim")
        kanal_listesi = kanal_ismi.split(",")
        for i in kanal_listesi:
            kumanda.kanal_ekle(i)
    elif islem == "5":
        print(f"Kanal sayısı: {len(kumanda)}")
    elif islem == "6":
        kumanda.rastgele_kanal()
    elif islem == "7":
        print(kumanda)
    else:
        print("Geçersiz bir işlem yaptınız !")
```
