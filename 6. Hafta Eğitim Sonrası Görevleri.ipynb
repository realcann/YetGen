Daha sonra, içinde çift ve tek sayılar bulunduran bir liste tanımlayın ve liste üzerinde gezinerek ekrana sadece çift sayıları bastırın.

```python
def cift(liste):
    for i in liste :
        if i% 2 in liste == 0 :
            print(i,end="")
        else :
            raise ValueError("Listede tek sayı var")
```
