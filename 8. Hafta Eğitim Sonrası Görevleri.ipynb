# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
def kullanıcıdan_isim_al():
    return input("İsim: ")

def kullanıcıdan_soyisim_al():
    return input("Soyisim: ")

def kullanıcıdan_yaş_al():
    return int(input("Yaş: "))

def kullanıcıdan_boy_al():
    return float(input("Boy (metre cinsinden): "))

def kullanıcıdan_kilo_al():
    return float(input("Kilo (kilogram cinsinden): "))

def vki_hesapla(kilo, boy):
    return kilo / (boy * boy)

def vki_sınıf_belirle(vki):
    if vki < 18.5:
        return "Zayıf"
    elif 18.5 <= vki <= 24.9:
        return "Normal kilolu"
    elif 25 <= vki <= 29.9:
        return "Fazla kilolu"
    elif 30 <= vki <= 39.9:
        return "Obez"
    else:
        return "İleri derecede obez"

def dosyaya_kaydet(isim, soyisim, yaş, boy, kilo, vki, sınıf):
    with open("veriler.txt", "a") as dosya:
        dosya.write(f"İsim: {isim}\n")
        dosya.write(f"Soyisim: {soyisim}\n")
        dosya.write(f"Yaş: {yaş}\n")
        dosya.write(f"Boy: {boy}\n")
        dosya.write(f"Kilo: {kilo}\n")
        dosya.write(f"VKİ: {vki}\n")
        dosya.write(f"Sınıf: {sınıf}\n")
        dosya.write("\n")

giris_sayisi = int(input("Kaç giriş yapacaksınız? "))

for _ in range(giris_sayisi):
    isim = kullanıcıdan_isim_al()
    soyisim = kullanıcıdan_soyisim_al()
    yaş = kullanıcıdan_yaş_al()
    boy = kullanıcıdan_boy_al()
    kilo = kullanıcıdan_kilo_al()

    vki = vki_hesapla(kilo, boy)
    sınıf = vki_sınıf_belirle(vki)

    dosyaya_kaydet(isim, soyisim, yaş, boy, kilo, vki, sınıf)
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
def menu_goruntule():
    print("YetBar - Smoothie Bank Menüsü:")
    print("1. Çilek Smoothie - 12 TL")
    print("2. Muz Smoothie - 10 TL")
    print("3. Şeftali Smoothie - 17 TL")
    print("4. Ananas Smoothie - 15 TL")
    print("5. Elma Smoothie - 7 TL")
    print("6. Vişne Smoothie - 13 TL")
def smoothie_satin_al():
    menu_goruntule()
    secim = int(input("Satın almak istediğiniz smoothie'nin numarasını girinİZ: "))
    adet = int(input("Kaç adet satın almak istiyorsunuz? "))

    if secim == 1:
        smoothie = "Çilek Smoothie"
        fiyat = 12
    elif secim == 2:
        smoothie = "Muzlu Smoothie"
        fiyat = 10
    elif secim == 3:
        smoothie = "Şeftali Smoothie"
        fiyat = 17
    elif secim == 4:
        smoothie = "Ananaslı Smoothie"
        fiyat = 15
    elif secim == 5:
        smoothie = "Elmalı Smoothie"
        fiyat = 7
    elif secim == 6:
        smoothie = "Vişneli Smoothie"
        fiyat = 13
    else:
        print("Geçersiz seçim yaptınız!")
        return

    toplam_fiyat = fiyat * adet

    # Satın alma işlemi loglanır
    log = f"{smoothie} - {adet} adet satın alındı. Toplam fiyat: {toplam_fiyat} TL"
    dosyaya_kaydet(log)

    print("Satın alma işlemi tamamlanmıştır.")
    print(f"Ödemeniz gereken toplam tutar: {toplam_fiyat} TL")

def dosyaya_kaydet(log):
    with open("smoothie_log.txt", "a") as dosya:
        dosya.write(log + "\n")
while True:
    print("YetBar - Smoothie Bank")
    print("1. Menüyü Görüntüle")
    print("2. Satın Alma İşlemleri")
    print("3. Çıkış")
    secim = int(input("Bir işlem seçin: "))

    if secim == 1:
        menu_goruntule()
    elif secim == 2:
        smoothie_satin_al()
    elif secim == 3:
        print("Programdan çıkılıyor...")
        break
    else:
        print("Geçersiz seçim!")
```
